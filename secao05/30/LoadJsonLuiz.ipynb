### Extracao

```python
dfAcoes = pd.read_json('tb_acoes.json', encoding='utf-8')
```

```python
dfAno = pd.read_json('tb_anos.json', encoding='utf-8')
```

```python
dfOrcamento = pd.read_json('tb_orcamentos.json', encoding='utf-8')
```

```python
dfPrograma = pd.read_json('tb_programas.json', encoding='utf-8')
```

```python
dfUni_orcamentaria = pd.read_json('tb_unidades_orcamentarias.json', encoding='utf-8')
```

```python
dfAcoes.head(1)
```

```python
dfOrcamento.head(1)
```

### Transformacao

```python
dfOrcamento = dfOrcamento.merge(dfAcoes,
                                left_on='fk_acao',
                                right_on='id_acao',
                                how='left')
```

```python
dfOrcamento.drop(columns=['fk_acao','id_acao'], inplace=True)
```

```python
dfPrograma.head(1)
```

```python
dfOrcamento = dfOrcamento.merge(dfPrograma,
                                left_on='fk_programa',
                                right_on='id_programa',
                                how='left').drop(columns=['fk_programa','id_programa'])
```

```python
dfOrcamento.head(1)
```

```python
dfAno.head(1)
```

```python
dfOrcamento = dfOrcamento.merge(dfAno,
                                left_on='fk_ano',
                                right_on='id_ano',
                                how='left').drop(columns=['fk_ano','id_ano'])
```

```python
dfOrcamento.head(2)
```

```python
dfUni_orcamentaria.head(1)
```

```python
dfOrcamento = dfOrcamento.merge(dfUni_orcamentaria,
                                left_on='fk_unidade',
                                right_on='id_unidade',
                                how='left').drop(columns=['fk_unidade','id_unidade'])
```

### Carregamento dos dados

```python
credenciais = service_account.Credentials.from_service_account_file(filename='chaveBigQuery.json',
                                                                    scopes=["https://www.googleapis.com/auth/cloud-platform"])
```

```python
dfOrcamento.to_gbq(credentials=credenciais,destination_table='cursoELTBasico.')
```
