### Transformacao

```python
# Total de atletas por modalidade e por pais

dfAtletas_disc = dfAtletas.groupby(['NOC', 'Discipline'])['Name'].nunique()
```

```python
dfAtletas_disc = pd.DataFrame(dfAtletas_disc).reset_index().rename(columns={'Name':'totalAtletas',
                                                                            'Discipline':'modalidade',
                                                                            'NOC':'pais'
                                                                           })
```

```python
dfTreinadores_disc = pd.DataFrame(dfTreinadores.groupby(['NOC', 'Discipline'])['Name'].nunique()). \
reset_index().rename(columns={'Name':'totalTreinadores',
                              'Discipline':'modalidade',
                              'NOC':'pais'
                              })
```

```python
dfTreinadores_disc
```

```python
dfTimes.head(20)
```

```python
dfTimes['categoria'] = dfTimes.Event.apply(lambda x: 'Masculino' if 'Men' in x
                                          else('Feminino' if 'Women' in x
                                              else ('Misto' if 'Mixed' in x
                                                   else 'Outros')))
```

```python
dfTimes
```

```python
dfTimes.categoria.value_counts()
```

```python
dfTimes_masc = pd.DataFrame(dfTimes[dfTimes.categoria=='Masculino'].groupby(['NOC', 'Discipline'])['Name'].nunique()). \
reset_index().rename(columns={'Name':'totalTimesMasculino',
                              'Discipline':'modalidade',
                              'NOC':'pais'
                              })
```

```python
dfTimes_fem = pd.DataFrame(dfTimes[dfTimes.categoria=='Feminino'].groupby(['NOC', 'Discipline'])['Name'].nunique()). \
reset_index().rename(columns={'Name':'totalTimeFeminino',
                              'Discipline':'modalidade',
                              'NOC':'pais'
                              })
```

```python
dfTimes_mis = pd.DataFrame(dfTimes[dfTimes.categoria=='Misto'].groupby(['NOC', 'Discipline'])['Name'].nunique()). \
reset_index().rename(columns={'Name':'totalTimeMisto',
                              'Discipline':'modalidade',
                              'NOC':'pais'
                              })
```

```python
dfTimes_out = pd.DataFrame(dfTimes[dfTimes.categoria=='Outros'].groupby(['NOC', 'Discipline'])['Name'].nunique()). \
reset_index().rename(columns={'Name':'totalTimeOutros',
                              'Discipline':'modalidade',
                              'NOC':'pais'
                              })
```

```python
dfTimes_masc
```

```python
dfTimes_fem
```

```python
dfTimes_mis
```

```python
### Juntando tabelas

lisDataFrames = [dfTimes_masc,dfTimes_fem,dfTimes_mis,dfTimes_out]
```

```python
df_final = reduce(lambda left, right: pd.merge(left,
                                               right,
                                               on=['pais', 'modalidade'],
                                              how='outer'), 
                  lisDataFrames).fillna(0)
```

```python
df_final
```

```python
df_final[df_final.pais == 'Brazil']
```

```python
##Levando os dados para o Bigquery

credenciais = service_account.Credentials.from_service_account_file(filename='chaveBigQuery.json'
                                                                    ,scopes=["https://www.googleapis.com/auth/cloud-platform"]
                                                                   )
```

```python
df_final.to_gbq(credentials=credenciais,
                       destination_table='cursoELTBasico.etl_excel',
                       if_exists='replace'
                       )
```
